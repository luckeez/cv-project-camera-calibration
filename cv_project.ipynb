```python
import numpy as np
import cv2

# Lista dei punti oggetto (3D)
object_points = np.array(world_coordinates, dtype=np.float32)

# Lista dei punti immagine (2D)
image_points = np.array(image_coordinates, dtype=np.float32)

# Dimensione dell'immagine
image_size = (1170, 658)


# Matrice della camera inizializzata come identità (verrà ottimizzata da calibrateCamera)
camera_matrix = np.eye(3, dtype=np.float32)

# Coefficienti di distorsione inizializzati a zero (verranno ottimizzati)
dist_coeffs = np.zeros((5,), dtype=np.float32)


# Variabili di output
rvecs = []  # Vettori di rotazione
tvecs = []  # Vettori di traslazione
```

```python
"""
Partendo dalla lista di punti costruisce `object_points` e `image_points` da passare al `calibrateCamera`
"""

def get_calibration_inputs(points_3d, points_2d):
    points_3d_array = np.array(points_3d)

    # Estrai i valori unici di z
    z_values = np.unique(points_3d_array[:, 2])

    # Crea una lista di array numpy, ognuno contenente i punti con lo stesso valore di z
    grouped_points_3d = [points_3d_array[points_3d_array[:, 2] == z_val] for z_val in z_values]

    # Raggruppa i punti 2D utilizzando la stessa struttura dei punti 3D
    grouped_points_2d = []
    start_idx = 0

    for z_val in z_values:
        # Conta il numero di punti per ciascun valore di z
        count = np.sum(points_3d_array[:, 2] == z_val)

        # Seleziona i punti corrispondenti nella lista dei punti 2D
        points_group = np.array(points_2d[start_idx:start_idx + count])

        # Modifica la struttura per avere la forma desiderata: (n_points, 1, 2)
        grouped_points_2d.append(points_group[:, np.newaxis, :].astype(np.float32))

        # Aggiorna l'indice di inizio
        start_idx += count

    # Convertiamo anche i punti 3D alla struttura desiderata: (n_points, 1, 3)
    grouped_points_3d = [group[:, np.newaxis, :].astype(np.float32) for group in grouped_points_3d]

    return grouped_points_3d, grouped_points_2d


object_points, image_points = get_calibration_inputs(world_coordinates[27:], image_coordinates[27:])

object_points
```

```python
# Esegui la calibrazione
ret, camera_matrix, dist_coeffs, rvecs, tvecs = cv2.calibrateCamera(
    object_points,      # Punti 3D nel mondo
    image_points,       # Punti 2D nell'immagine
    image_size,         # Dimensione dell'immagine
    camera_matrix,      # Matrice della camera (inizialmente stimata, verrà ottimizzata)
    dist_coeffs,        # Coefficienti di distorsione
    flags=cv2.CALIB_USE_INTRINSIC_GUESS # Tell the function to use the provided initial camera matrix
)

print(ret)
print(camera_matrix)
print(dist_coeffs)
print(rvecs)
print(tvecs)
```

```python
 # EVALUATION

print("Punti proiettati sull'immagine:")
errors = []
for i in range(len(object_points)):

  # Proietta i punti 3D sull'immagine utilizzando i parametri calcolati
  projected_point, _ = cv2.projectPoints(
      object_points[i],  # Punti 3D
      rvecs[0],          # Vettore di rotazione
      tvecs[0],          # Vettore di traslazione
      camera_matrix,     # Matrice della camera
      dist_coeffs        # Coefficienti di distorsione
  )
  projected_point = projected_point.squeeze()


  print(f"real result: [{image_points[i][0]}, {image_points[i][1]}]")
  print(f"calc result: [{projected_point[0]}, {projected_point[1]}]\n")

  error = (image_points[i][0]-projected_point[0])**2 + (image_points[i][1]-projected_point[1])**2
  errors.append(error)

print("average error: ", sum(errors)/len(errors))
```

```python
def project_3dTo2d(transformation_matrix, point_3d):
  point_3d = np.append(np.array(point_3d), 1)
  result = np.dot(transformation_matrix, point_3d)
  ic = [result[0]/result[2], result[1]/result[2]]
  return ic
```

```python
def construct_linear_system(image_coordinates, world_coordinates):
  A = []
  for i in range(len(image_coordinates)):
    wc = world_coordinates[i]
    ic = image_coordinates[i]
    a1 = [wc[0], wc[1], wc[2], 1, 0, 0, 0, 0, -wc[0]*ic[0], -wc[1]*ic[0], -wc[2]*ic[0]]
    a2 = [0, 0, 0, 0, wc[0], wc[1], wc[2], 1, -wc[0]*ic[1], -wc[1]*ic[1], -wc[2]*ic[1]]
    A.append(a1)
    A.append(a2)

  b = []
  for i in range(len(image_coordinates)):
    b.append(image_coordinates[i][0])
    b.append(image_coordinates[i][1])

  A = np.array(A)
  b = np.array(b)

  return A, b
```
